```python
def count_leq(x, array):
    left = -1
    right = len(array)
    while left + 1 < right:
        mid = (left + right) / 2
        if array[mid] <= x:
            right = mid
        else:
            left = mid
    return float(len(array) - right) / len(array)

figure()
plot(arange(0, 25, 0.01), [count_leq(x, kld) for x in arange(0, 25, 0.01)], 'r')
xlabel('KL Divergence')
ylabel('CDF')
title('KL Divergence')
show()
```

```python
print len(rank)
sample_size = 20
print "# Top", sample_size, "Words #"
for k in range(sample_size):
    print rank_words[k], rank[k]
print
print "# Last", sample_size, "Words #"
for k in range(-sample_size, 0):
    print rank_words[k], rank[k]
```
